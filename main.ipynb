```python
# Use the csv module to load data from a CSV file into memory
import csv
import pandas as pd
import tui
import process



def read_data(file_name):
  tui.started(f"Reading data from {file_name}")

  data = []

  with open(file_name, encoding="UTF-8") as csv_file:
    # Read the CSV file using the csv.reader() function.
    csv_reader = csv.reader(csv_file)
    headings = next(csv_reader)

    # Store the data in a list (do not convert it to a DataFrame).
    for row in csv_reader:
        data.append(row)

  # Load data from a CSV file into a Pandas DataFrame using the read_csv() function.
  # Use the file path or filename obtained from Task A (do not prompt the user again).
  df = pd.read_csv(file_name, encoding='utf-8', index_col='Patient_ID')

  tui.completed()
  return data, headings, df



def run():
  # Prompt the user to input the file path or filename for the CSV file.
  #file_name = input("Input CSV file path/name: ") # lung_cancer_data.csv
  file_name = "lung_cancer_data.csv" # TEST!
  data, headings, df = read_data(file_name)

  while True: # loop - returns back to the menu
    # Display the menu
    selected_option = tui.display_menu()

    ### CHECK: A4, C4

    if selected_option == '1':
      process.A1(data) # test : 13117

    elif selected_option == '2':
      process.A2(data) # test : Hispanic, Asian

    elif selected_option == '3':
      process.A3(data) # test : Surgery

    elif selected_option == '4':
      process.A4(data, headings)

    elif selected_option == '5':
      process.B1(df)

    elif selected_option == '6':
      process.B2(df)

    elif selected_option == '7':
      process.B3(df)

    elif selected_option == '8':
      process.B4(df)

    elif selected_option == '9':
      process.C1(df)

    elif selected_option == '10':
      process.C2(df)

    elif selected_option == '11':
      process.C3(df)

    elif selected_option == '12':
      process.C4(df)

    elif selected_option == '13':
      break

    else:
      tui.error("Invalid Selection!")



try:
    if __name__ == "__main__":
        run()
except FileNotFoundError:
    print("File does not exist")

### CHANGE back all TEST! codes
```
